```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

np.random.seed(0)
x = np.arange(-1,1,0.1)
y = -x**4+2*x**2 + np.random.normal(0,0.1,len(x))

X_train,X_test,y_train,y_test = train_test_split(x,y,train_size = 0.7)



for degree in [1,4,11]:
    p = np.poly1d(np.polyfit(X_train,y_train,degree))
    plt.scatter(X_train,y_train,label="Valeurs d'entraînement")
    plt.scatter(X_test,y_test,label="Valeurs de test")  
    curve_x = np.arange(min(x),max(x),0.01)
    plt.plot(curve_x,p(curve_x),label="Polynome de degré "+ str(degree))
    plt.xlim(-1,1)
    plt.ylim(-0.5,1.5)
    plt.legend()  
    plt.show()
```
